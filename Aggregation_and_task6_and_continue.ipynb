```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient["Summary_Hashtag_Sentiment_Week"]
collection_list = mydb.list_collection_names()
collection_list.sort()
print(collection_list)
```

```python
weeks=[]
for i in (collection_list):
    if(i[-9:]=="Sentiment"):
        if(i[0]=="2"):
            weeks.append(i)
print(weeks)
#print(len(weeks))
```

```python
for week in weeks:
    print(week)
    distribution_neutral = "distribution_neutral-"+week
    distribution_positive = "distribution_positive-"+week
    distribution_negative = "distribution_negative-"+week
    mycol = mydb[week]
    pipeline = [
    { '$group' : {'_id': { '$toLower' :"$Hashtag"}, 'Logits_Neutral':{'$sum': '$Logits_Neutral'}}},
    { '$sort'  : {'Logits_Neutral':-1 }},
    { '$out' : distribution_neutral}
    ]
    mycol.aggregate(pipeline,allowDiskUse=True)
    
for week in weeks:
    print(week)
    distribution_neutral = "distribution_neutral-"+week
    distribution_positive = "distribution_positive-"+week
    distribution_negative = "distribution_negative-"+week
    mycol = mydb[week]
    pipeline = [
    { '$group' : {'_id': { '$toLower' :"$Hashtag"}, 'Logits_Positive':{'$sum': '$Logits_Positive'}}},
    { '$sort'  : {'Logits_Positive':-1 }},
    { '$out' : distribution_positive}
    ]
    mycol.aggregate(pipeline,allowDiskUse=True)
    
for week in weeks:
    print(week)
    distribution_neutral = "distribution_neutral-"+week
    distribution_positive = "distribution_positive-"+week
    distribution_negative = "distribution_negative-"+week
    mycol = mydb[week]
    pipeline = [
    { '$group' : {'_id': { '$toLower' :"$Hashtag"}, 'Logits_Negative':{'$sum': '$Logits_Negative'}}},
    { '$sort'  : {'Logits_Negative':-1 }},
    { '$out' : distribution_negative}
    ]
    mycol.aggregate(pipeline,allowDiskUse=True)
```
