```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
dblist = myclient.list_database_names()
mydb = myclient["Summary_Hashtag_Sentiment_Week"]
```

```python
def func_divide_week(listTemp, n):
    for i in range(0, len(listTemp), n):
        yield listTemp[i:i + n]
```

```python
import pandas as pd
import os
import random
Start_date = '2020-01-22'
End_date = '2021-03-30'
datelist = pd.date_range(Start_date, End_date).tolist()
temp = func_divide_week(datelist, 7)
first = "#covid19"
second = "#coronavirus"
third = "#covid"
forth = "#corona"
fifth = "#lockdown"
list_Hashtag_first = []
list_Hashtag_second= []
list_Hashtag_third= []
list_Hashtag_forth= []
list_Hashtag_fifth= []
for week in temp:
    collection_name = "distribution_neutral-"+str(week[0])[0:4]+"_"+str(week[0])[5:7]+"_"+str(week[0])[8:10]+"_Week_Hashtag_Sentiment"
    mycol = mydb[collection_name]
    week_num = mycol.find({"_id":first})
    if(week_num.count()==0):
        list_Hashtag_first.append(int(0))
    if(week_num.count()!=0):
        for x in week_num:
            list_Hashtag_first.append(x['Logits_Neutral'])
    week_num = mycol.find({"_id":second})
    if(week_num.count()==0):
        list_Hashtag_second.append(int(0))
    if(week_num.count()!=0):
        for x in week_num:
            list_Hashtag_second.append(x['Logits_Neutral'])
    week_num = mycol.find({"_id":third})
    if(week_num.count()==0):
        list_Hashtag_third.append(int(0))
    if(week_num.count()!=0):
        for x in week_num:
            list_Hashtag_third.append(x['Logits_Neutral'])
    week_num = mycol.find({"_id":forth})
    if(week_num.count()==0):
        list_Hashtag_forth.append(int(0))
    if(week_num.count()!=0):
        for x in week_num:
            list_Hashtag_forth.append(x['Logits_Neutral'])
    week_num = mycol.find({"_id":fifth})
    if(week_num.count()==0):
        list_Hashtag_fifth.append(int(0))
    if(week_num.count()!=0):
        for x in week_num:
            list_Hashtag_fifth.append(x['Logits_Neutral'])
print(list_Hashtag_first)
print(list_Hashtag_second)
print(list_Hashtag_third)
print(list_Hashtag_forth)
print(list_Hashtag_fifth)
```
