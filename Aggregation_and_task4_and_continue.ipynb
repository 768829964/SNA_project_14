```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient["Summary_Hashtag_Details_Week"]
collection_list = mydb.list_collection_names()
collection_list.sort()
print(collection_list)
```

```python
weeks=[]
for i in (collection_list):
    if(i[-7:]=="Details"):
        if(i[0]=="2"):
            weeks.append(i)
print(weeks)
#print(len(weeks))
```

```python
for week in weeks:
    print(week)
    distribution_reply = "distribution_reply-"+week
    distribution_like = "distribution_like-"+week
    mycol = mydb[week]
    pipeline = [
    { '$group' : {'_id': { '$toLower' :"$Hashtag"}, 'likenum':{'$sum': '$Likes'}}},
    { '$sort'  : {'likenum':-1 }},
    { '$out' : distribution_like}
    ]
    mycol.aggregate(pipeline,allowDiskUse=True)
    
```

```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient["Summary_Hashtag_Details_Week"]
mycol = mydb["distribution_all_like"]
pipeline = [
    { '$group' : {'_id': '$Hashtag', 'likenum':{'$sum': '$likenum'}}},
    { '$sort'  : {'likenum':-1 }},
    { '$out' : 'distribution_final_like'}
]
mycol.aggregate(pipeline, allowDiskUse=True)
```
