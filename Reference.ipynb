```python
def func_divide_week(listTemp, n):
    for i in range(0, len(listTemp), n):
        yield listTemp[i:i + n]
```

```python
import pandas as pd
Hashtag_first_1 = "#covid19"
Hashtag_first_2 = "#COVID19"
Hashtag_first_3 = "#Covid19"
Start_date = '2020-1-22'
End_date = '2020-9-15'
datelist = pd.date_range(Start_date, End_date).tolist()
temp = func_divide_week(datelist, 7)
list_Hashtag_first = []
for week in temp:
    week_num = 0
    for date in week:
        day = date.strftime("%Y_%m_%d")
        collection_name = day+"_Summary_Hashtag"
        mycol = mydb[collection_name]
        week_num = week_num+mycol.find({'$or':[{"Hashtag":Hashtag_first_1},{"Hashtag":Hashtag_first_2},{"Hashtag":Hashtag_first_3}]}).count()
    print(week_num)
    list_Hashtag_first.append(week_num)
    
Start_date = '2020-9-16'
End_date = '2021-3-31'
datelist = pd.date_range(Start_date, End_date).tolist()
temp = func_divide_week(datelist, 7)
list_Hashtag_first_1 = []
for week in temp:
    week_num = 0
    for date in week:
        day = date.strftime("%Y_%m_%d")
        collection_name = day+"_Summary_Hashtag"
        mycol = mydb[collection_name]
        week_num = week_num+mycol.find({'$or':[{"Hastag":Hashtag_first_1},{"Hastag":Hashtag_first_2},{"Hastag":Hashtag_first_3}]}).count()
    print(week_num)
    list_Hashtag_first_1.append(week_num)
    
list_Hashtag_first_2 = list_Hashtag_first[0:34]
list_Hashtag_first_2.append(list_Hashtag_first[34]+list_Hashtag_first_1[0])
list_Hashtag_first_2.extend(list_Hashtag_first_1[1:])
print(list_Hashtag_first_2)
```
