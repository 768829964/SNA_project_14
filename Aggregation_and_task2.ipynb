```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient["Summary_NER_Week"]
collection_list = mydb.list_collection_names()
collection_list.sort()
print(collection_list)
```

```python
for week in collection_list:
    print(week)
    mycol = mydb[week]
    pipeline = [
    { '$group':{'_id': { '$toLower' :"$NER_Text"}, 'num': {'$sum': 1 }}},
    { '$out':'distribution-'+week}
    ]
    mycol.aggregate(pipeline)
```

```python
from pymongo import MongoClient
import pymongo

myclient = pymongo.MongoClient('mongodb://localhost:27017/')
mydb = myclient["Summary_NER_Week"]
mycol = mydb["distribution_all"]
pipeline = [
    { '$group' : {'_id': '$NER_Text', 'totalnum':{'$sum': '$num'}}},
    { '$sort'  : {'totalnum':-1 }},
    { '$out' : 'distribution_final'}
]
mycol.aggregate(pipeline, allowDiskUse=True)
```
